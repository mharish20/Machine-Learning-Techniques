```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from math import sqrt
```

```python
# 1. Read CSV data into pandas DataFrame object
df = pd.read_csv("LinearRegression.csv")  
```

```python
# 2. Do necessary preprocessing (for example, drop missing values)
df = df.dropna()
```

```python
# 3. Choose independent variable (X) and dependent variable (Y)
# Assume the CSV has columns 'X' and 'Y'
X = df['X'].values
Y = df['Y'].values
```

```python
# 4. Find the b0 and b1 values for Y_predicted using simple formulas:
#    b1 = Σ((X - mean(X))*(Y - mean(Y))) / Σ((X - mean(X))^2)
#    b0 = mean(Y) - b1*mean(X)
mean_x = np.mean(X)
mean_y = np.mean(Y)
b1 = np.sum((X - mean_x) * (Y - mean_y)) / np.sum((X - mean_x)**2)
b0 = mean_y - b1 * mean_x

# Predicted Y
Y_pred = b0 + b1 * X
```

```python
# 5. Calculate the SSE (sum of squared error)
SSE = np.sum((Y - Y_pred)**2)
print("SSE:", SSE)
```

```python
# 6. Calculate the RMSE (Root Mean Square Error) value
rmse = sqrt(SSE/len(Y))
print("RMSE:", rmse)
```

```python
# 7. Calculate the coefficient of determination (r^2)
SST = np.sum((Y - mean_y)**2)
r2 = 1 - (SSE/SST)
print("R^2:", r2)
```

```python
# 8. Plot regression line along with the given data points
plt.scatter(X, Y, color='blue', label='Data Points')
plt.plot(X, Y_pred, color='red', label='Regression Line')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Linear Regression')
plt.legend()
plt.show()
```

```python
# 9. Predict the output for a given input value
input_value = 10  # Change this value for prediction
predicted_value = b0 + b1 * input_value
print(f"Predicted Y value for X = {input_value} is {predicted_value}")
```
