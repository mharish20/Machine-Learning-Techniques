```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
# 1. Consider the given dataset and 2. Do necessary preprocessing
df = pd.read_csv("MultiLinearRegression.csv")  # Replace with your CSV file name
df = df.dropna()
```

```python
# 3. Choose independent variables and dependent variable from given dataset
# Assume 'Y' is the dependent variable and others (e.g., 'X1', 'X2') are independents
feature_columns = ['X1', 'X2']  # Adjust as needed
X = df[feature_columns].values
Y = df['Y'].values
```

```python
# 4. Print values of y-intercept and independent variable coefficients
model = LinearRegression()
model.fit(X, Y)
intercept = model.intercept_
coefficients = model.coef_
print("Y-intercept:", intercept)
print("Coefficients:", coefficients)
```

```python
# 5. Find the Y_pred
Y_pred = model.predict(X)
```

```python
# 6. Calculate the SSE and RMSE (Root Mean Square Error) value
SSE = np.sum((Y - Y_pred)**2)
rmse = np.sqrt(mean_squared_error(Y, Y_pred))
print("SSE:", SSE)
print("RMSE:", rmse)
```

```python
# 7. Calculate the coefficient of determination (r^2)
r2 = model.score(X, Y)
print("R^2:", r2)
```

```python
# 8. Plot the scatter graph of Y_actual and Y_predicted
plt.scatter(Y, Y_pred, color='green')
plt.xlabel("Actual Y")
plt.ylabel("Predicted Y")
plt.title("Actual vs Predicted Y")
plt.plot([min(Y), max(Y)], [min(Y), max(Y)], color='red')  # Reference line for perfect prediction
plt.show()
```

```python
# 9. Predict the output for given input values
# For example, predicting for X1=5 and X2=3. Adjust as needed.
input_values = [[5, 3]]
predicted_value = model.predict(input_values)
print(f"Predicted Y value for input {input_values[0]} is {predicted_value[0]}")
```
